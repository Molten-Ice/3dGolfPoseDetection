```python
import math
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, writers
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import subprocess as sp

def get_resolution(filename):
    """Returns height, width of video"""
    command = ['ffprobe', '-v', 'error', '-select_streams', 'v:0',
               '-show_entries', 'stream=width,height', '-of', 'csv=p=0', filename]
    with sp.Popen(command, stdout=sp.PIPE, bufsize=-1) as pipe:
        for line in pipe.stdout:
            w, h = line.decode().strip().split(',')
            return int(w), int(h)
            
def get_fps(filename):
    """Returns the fps of the video"""
    command = ['ffprobe', '-v', 'error', '-select_streams', 'v:0',
               '-show_entries', 'stream=r_frame_rate', '-of', 'csv=p=0', filename]
    with sp.Popen(command, stdout=sp.PIPE, bufsize=-1) as pipe:
        for line in pipe.stdout:
            a, b = line.decode().strip().split('/')
            return int(a) / int(b)

def read_video(filename, skip=0, limit=-1):
    """This function reads a video file and yields each frame as a numpy array in RGB format"""
    w, h = get_resolution(filename)
    
    command = ['ffmpeg',
            '-i', filename,
            '-f', 'image2pipe',
            '-pix_fmt', 'rgb24',
            '-vsync', '0',
            '-vcodec', 'rawvideo', '-']
    
    i = 0
    with sp.Popen(command, stdout = sp.PIPE, bufsize=-1) as pipe:
        while True:
            data = pipe.stdout.read(w*h*3)
            if not data:
                break
            i += 1
            if i > limit and limit != -1:
                continue
            if i > skip:
                yield np.frombuffer(data, dtype='uint8').reshape((h, w, 3))

def dot_product(a, b):
    return a[0] * b[0] + a[1] * b[1]

def magnitude(vector):
    return math.sqrt(vector[0]**2 + vector[1]**2)

def angle_between_vectors(a, b):
    if len(a) != 2 or len(b) != 2:
        raise ValueError("Both vectors must have exactly 2 dimensions")

    dot_product_ab = dot_product(a, b)
    magnitude_a = magnitude(a)
    magnitude_b = magnitude(b)

    cos_angle = dot_product_ab / (magnitude_a * magnitude_b)
    angle_rad = math.acos(cos_angle)
    angle_deg = math.degrees(angle_rad)

    return angle_deg

def euclidean_distance_2d(point1, point2):
    if len(point1) != 2 or len(point2) != 2:
        raise ValueError("Points must have 2 dimensions")

    distance = ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5
    return distance

def rotate_points(points, pivot, angle_degrees):
    angle_radians = math.radians(angle_degrees)
    rotated_points = []

    for point in points:
        # Translate the point so that the pivot point becomes the origin
        translated_x = point[0] - pivot[0]
        translated_y = point[1] - pivot[1]

        # Rotate the translated point around the origin
        rotated_x = translated_x * math.cos(angle_radians) - translated_y * math.sin(angle_radians)
        rotated_y = translated_x * math.sin(angle_radians) + translated_y * math.cos(angle_radians)

        # Translate the rotated point back to the original coordinate system
        final_x = rotated_x + pivot[0]
        final_y = rotated_y + pivot[1]

        rotated_points.append((final_x, final_y))

    return np.array(rotated_points)
```

```python
import numpy as np

def normalize_vector(v):
    return v / np.linalg.norm(v)

def angle_between_lines(line1_point1, line1_point2, line2_point2):
    line2_point1 = line1_point1
    # Calculate direction vectors for each line
    direction_vector1 = normalize_vector(np.array(line1_point2) - np.array(line1_point1))
    direction_vector2 = normalize_vector(np.array(line2_point2) - np.array(line2_point1))

    # Calculate the dot product between direction vectors
    dot_product = np.dot(direction_vector1, direction_vector2)

    # Calculate the angle between the lines using the dot product and direction vector magnitudes
    angle_rad = np.arccos(np.clip(dot_product, -1, 1))  # Clip to handle potential floating point errors
    angle_deg = np.degrees(angle_rad)

    return angle_deg

# Use line1_point1 as the first point of the second line
angle_degrees = angle_between_lines(red_wrist_3d, red_foot_3d, black_foot_3d)
print("Angle between the two lines:", angle_degrees, "degrees")

import numpy as np

def normalize_vector(v):
    return v / np.linalg.norm(v)

def rodrigues_rotation(points, line_point1, line_point2, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    axis = normalize_vector(np.array(line_point2) - np.array(line_point1))

    rotated_points = []
    for point in points:
        point = np.array(point)
        line_point1 = np.array(line_point1)

        # Translate the point so that line_point1 becomes the origin
        translated_point = point - line_point1

        # Apply the Rodrigues' rotation formula
        rotated_point = (translated_point * np.cos(angle_radians)
                         + np.cross(axis, translated_point) * np.sin(angle_radians)
                         + axis * np.dot(axis, translated_point) * (1 - np.cos(angle_radians)))

        # Translate the rotated point back to the original coordinate system
        final_point = rotated_point + line_point1

        rotated_points.append(tuple(final_point))

    return np.array(rotated_points)

# Example usage:
pos = rodrigues_rotation(pos, red_wrist_3d, red_foot_3d, -angle_degrees)
print("Rotated points:", pos)

red_foot_3d, black_foot_3d, red_wrist_3d = pos[[3, 6, 16]]


print(f"3D after second rotation of : {angle_degrees} degress |", red_foot_3d, black_foot_3d, red_wrist_3d)
print()
differences = pos[valid_pairings[:, 0], :2] -  frame_2d_keypoints[valid_pairings[:, 1]]
print(f"Average difference: {abs(differences).mean():.3f}")
# [740.67554 923.6686  -57.64646] [1026.4302   942.80664  -49.24027] [ 908.73376  594.1759  -166.00035]
```
